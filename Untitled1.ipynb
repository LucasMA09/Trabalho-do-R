```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pathlib
import os
import zipfile
import seaborn as sns
```

```python
import pandas as pd

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

df = pd.read_csv("https://raw.githubusercontent.com/drobaina/Python/main/microdados_ed_basica_RJ_2022.csv", header=0)
df.head()
```

```python
dados_nova_frib = df[df["NO_MUNICIPIO"] == "Rio de Janeiro"]

dados_nova_frib.head()
```

```python
# IN_LIXO_SERVICO_COLETA: se realiza ou não a coleta de lixo
# IN_ESGOTO_REDE_PUBLICA: se tem ou não rede pública de esgoto

# Agrupar os dados
nova_base_de_dados = dados_nova_frib[["IN_LIXO_SERVICO_COLETA", "IN_ESGOTO_REDE_PUBLICA"]]

nova_base_de_dados.head()
```

```python
ocorrencia_de_coleta = dados_nova_frib.groupby('IN_LIXO_SERVICO_COLETA').size().reset_index(name='numero_de_ocorrencias')

ocorrencia_de_coleta
```

```python
ocorrencia_de_rede = dados_nova_frib.groupby('IN_ESGOTO_REDE_PUBLICA').size().reset_index(name='numero_de_ocorrencias')

ocorrencia_de_rede
```

```python
### gráfico de colunas (coleta de lixo)
plt.figure(figsize=(14, 6))
plt.bar(ocorrencia_de_coleta['IN_LIXO_SERVICO_COLETA'], ocorrencia_de_coleta['numero_de_ocorrencias'])
plt.title('Coleta de lixo')
plt.xlabel('Se tem ou não')
plt.ylabel('Número de Ocorrências')
plt.xticks(ocorrencia_de_coleta['IN_LIXO_SERVICO_COLETA'])
plt.tight_layout()
plt.show()
```

```python
### gráfico de colunas (rede pública de esgoto)
plt.figure(figsize=(14, 6))
plt.bar(ocorrencia_de_rede['IN_ESGOTO_REDE_PUBLICA'], ocorrencia_de_rede['numero_de_ocorrencias'])
plt.title('Rede Pública de Esgoto')
plt.xlabel('Se tem ou não')
plt.ylabel('Número de Ocorrências')
plt.xticks(ocorrencia_de_rede['IN_ESGOTO_REDE_PUBLICA'])
plt.tight_layout()
plt.show()
```

Era do entendiemnto do grupo que rede pública de esgoto tende a ser mais valorizada em meios urbanos, e, frente a outros metodos de descarte de dejetos, a sua implementação seria priorizada. No entanto, ao analisarmos os dados do município do Rio de Janeiro fica evidente que esse não é o caso, visto que, dentre os 3929 agentes analisados 15 não apresentam rede publica de esgoto, enquanto apenas 12 não realizam a coleta de lixo, portanto fica claro qeu a coleta é mais valorizada.
