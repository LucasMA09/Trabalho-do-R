```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pathlib
import os
import zipfile
import seaborn as sns
```

```python
import pandas as pd

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

df = pd.read_csv("https://raw.githubusercontent.com/drobaina/Python/main/microdados_ed_basica_RJ_2022.csv", header=0)
df.head()
```

```python
dados_nova_frib = df[df["NO_MUNICIPIO"] == "Rio de Janeiro"]

dados_nova_frib.head()
```

```python
# IN_LIXO_SERVICO_COLETA: se realiza ou n√£o a coleta de lixo
# IN_ESGOTO_REDE_PUBLICA: se tem ou n√£o rede p√∫blica de esgoto

# Agrupar os dados
nova_base_de_dados = dados_nova_frib[["IN_LIXO_SERVICO_COLETA", "IN_ESGOTO_REDE_PUBLICA"]]

nova_base_de_dados.head()
```

```python
ocorrencia_de_coleta = dados_nova_frib.groupby('IN_LIXO_SERVICO_COLETA').size().reset_index(name='numero_de_ocorrencias')

ocorrencia_de_coleta
```

```python
ocorrencia_de_rede = dados_nova_frib.groupby('IN_ESGOTO_REDE_PUBLICA').size().reset_index(name='numero_de_ocorrencias')

ocorrencia_de_rede
```

```python
### gr√°fico de colunas (coleta de lixo)
plt.figure(figsize=(14, 6))
plt.bar(ocorrencia_de_coleta['IN_LIXO_SERVICO_COLETA'], ocorrencia_de_coleta['numero_de_ocorrencias'])
plt.title('Coleta de lixo')
plt.xlabel('Se tem ou n√£o')
plt.ylabel('N√∫mero de Ocorr√™ncias')
plt.xticks(ocorrencia_de_coleta['IN_LIXO_SERVICO_COLETA'])
plt.tight_layout()
plt.show()
```

```python
### gr√°fico de colunas (rede p√∫blica de esgoto)
plt.figure(figsize=(14, 6))
plt.bar(ocorrencia_de_rede['IN_ESGOTO_REDE_PUBLICA'], ocorrencia_de_rede['numero_de_ocorrencias'])
plt.title('Rede P√∫blica de Esgoto')
plt.xlabel('Se tem ou n√£o')
plt.ylabel('N√∫mero de Ocorr√™ncias')
plt.xticks(ocorrencia_de_rede['IN_ESGOTO_REDE_PUBLICA'])
plt.tight_layout()
plt.show()
```
